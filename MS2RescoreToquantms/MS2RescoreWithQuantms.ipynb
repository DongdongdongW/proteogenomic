```python
before_protein = set(before_ms2rescore.protein_table["accession"].to_list())
before_peptide = set(before_ms2rescore.peptide_table["opt_global_cv_MS:1000889_peptidoform_sequence"].to_list())
before_psm = set(before_ms2rescore.spectrum_match_table["opt_global_cv_MS:1000889_peptidoform_sequence"].to_list())

after_protein = set(after_ms2rescore.protein_table["accession"].to_list())
after_peptide = set(after_ms2rescore.peptide_table["opt_global_cv_MS:1000889_peptidoform_sequence"].to_list())
after_psm = set(after_ms2rescore.spectrum_match_table["opt_global_cv_MS:1000889_peptidoform_sequence"].to_list())
```

```python
x = ['protein', 'peptide', 'PSM']
before = [len(before_protein), len(before_peptide),len(before_psm)]
after = [len(after_protein), len(after_peptide), len(after_psm)]

bar_width = 0.2
r1 = range(len(x))
r2 = [x + bar_width for x in r1]
plt.bar(r1, before, width=bar_width, label='Before')
plt.bar(r2, after, width=bar_width, label='After')
for i, value in enumerate(before):
    plt.text(i, value, str(value), ha='center', va='bottom')

for i, value in enumerate(after):
    plt.text(i + bar_width, value, str(value), ha='center', va='bottom')
plt.xticks([r + bar_width / 2 for r in range(len(x))], x)

plt.title('Results before and after using MS2Rescore')
plt.ylabel('Number')
plt.legend()
```

```python
venn = venn2([before_protein,after_protein], set_labels=('before_protein', 'after_protein'))
venn.get_label_by_id('10').set_text(len(before_protein - after_protein))
venn.get_label_by_id('01').set_text(len(after_protein - before_protein))
venn.get_label_by_id('11').set_text(len(before_protein & after_protein))
plt.title('Protein before and after using MS2Rescore')
plt.show()
```

```python
venn = venn2([before_peptide,after_peptide], set_labels=('before_peptide', 'after_peptide'))
venn.get_label_by_id('10').set_text(len(before_peptide - after_peptide))
venn.get_label_by_id('01').set_text(len(after_peptide - before_peptide))
venn.get_label_by_id('11').set_text(len(before_peptide & after_peptide))
plt.title('Peptide before and after using MS2Rescore')
plt.show()
```

```python
venn = venn2([before_psm,after_psm], set_labels=('before_psm', 'after_psm'))
venn.get_label_by_id('10').set_text(len(before_psm - after_psm))
venn.get_label_by_id('01').set_text(len(after_psm - before_psm))
venn.get_label_by_id('11').set_text(len(before_psm & after_psm))
plt.title('PSM before and after using MS2Rescore')
plt.show()
```
